```python
heart_d=pd.read_csv('https://raw.githubusercontent.com/mrdbourke/zero-to-mastery-ml/master/data/heart-disease.csv')
heart_d.head()
```

```python
heart_d['age'].plot.hist();
```

```python
over50=heart_d[heart_d['age']>50]
over50.head()
```

```python
over50.plot(kind='scatter',x='age',y='chol',c='target');
```

```python
fig,ax=plt.subplots(figsize=(10,6))

over50.plot(kind='scatter',x='age',y='chol',c='target',ax=ax)
```

```python
fig,ax=plt.subplots(figsize=(10,6))

over50.plot(kind='scatter',x='age',y='chol',c='target',ax=ax)
```

```python
fig,ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(x=over50['age'],y=over50['chol'],c=over50['target'])
ax.set(title='heart diease and cholesterol levels',xlabel='age',ylabel='cholesteral')

ax.legend(*scatter.legend_elements(),title='Target')

ax.axhline(over50['chol'].mean(),linestyle='--')
```

```python
plt.style.use('seaborn-whitegrid')

fig,ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(x=over50['age'],y=over50['chol'],c=over50['target'],cmap='winter')
ax.set(title='heart diease and cholesterol levels',xlabel='age',ylabel='cholesteral')

ax.legend(*scatter.legend_elements(),title='Target')

ax.axhline(over50['chol'].mean(),linestyle='--')
```

```python
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)
fig.suptitle('Heart Disease Analysis',fontsize=20,fontweight='bold')

ax0.scatter(x=over50.age,y=over50.chol,c=over50.target,cmap='winter')
ax0.set(title='heart disease and cholesterol levels',ylabel='Cholesterol',xlim=[50,80])
ax0.legend(*scatter.legend_elements(),title='Target')
ax0.axhline(y=over50.chol.mean(),linestyle='--',c='red')

ax1.scatter(x=over50.age,y=over50.thalach,c=over50.target,cmap='winter')
ax1.set(title='heart disease and thalach levels',xlabel='Age',ylabel='Thalach',ylim=[60,200])
ax1.legend(*scatter.legend_elements(),title='Target')
ax1.axhline(over50.thalach.mean(),ls=':',c='red')
```
